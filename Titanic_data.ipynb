```python
#to import the various modules

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df= sns.load_dataset("titanic")

#print(titanic.head())
# Display the rows and column of the dataset
df.head()
```

```python
#To get all the column/ features .
#using the df.column to get all the column .
#The dataset contains information about passengers (age, gender, class, etc.).
df.columns
```

```python
# To get the missing data set for analysis 
df.isnull().sum()
```

```python
# Fill missing age values with the median
# Fill missing age values with the median
print(df.columns)
```

```python
# Ensure 'Unknown' is added as a category before filling missing values
df['deck'] = df['deck'].astype('category')  # Ensure it's categorical
df['deck'] = df['deck'].cat.add_categories(['Unknown'])  # Add 'Unknown' as a valid category
df['deck'] = df['deck'].fillna('Unknown')  # Now fill missing values


```

```python
sns.countplot(x="sex", hue="survived", data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
sns.barplot(x="class", y="survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```
